*prompt utilizado*

"Un príncipe danés de mirada melancólica, vestido con ropajes oscuros, sostiene una calavera en su mano. Su expresión refleja profunda contemplación y angustia existencial. La escena es sombría, iluminada por una tenue luz azulada que resalta las sombras de su rostro.
```python
prompt = "A melancholic Danish prince in dark clothing holds a skull in his hand. His expression shows deep contemplation and existential anguish. The scene is somber, illuminated by a faint bluish light that highlights the shadows of his face. The background reveals a medieval castle, shrouded in mysterious fog, evoking a sense of introspection and tragic fate."
```

```python
url = "https://api.stability.ai/v2beta/stable-image/generate/ultra"
headers = {
    "authorization": f"Bearer {API_KEY}",
    "accept": "image/*"
}
params = {
    "prompt": prompt,
    "output_format": "png"
}
```

```python
response = requests.post(
    url,
    headers=headers,
    files={"none": ''},
    data=params
)
```


***3. Aplicación y Ensayo ***

- Usa imágenes generadas para reforzar la memoria visual.

```python

print(response.status_code)
if response.status_code == 200:

    with open("imagen.png", 'wb') as file:
        file.write(response.content)
    print("✅ Imagen guardada como 'imagen.png'")


    from IPython.display import Image
    display(Image(response.content))
else:
    error_msg = response.json().get("message", "Error desconocido")
    raise Exception(f"❌ Error {response.status_code}: {error_msg}")
```
