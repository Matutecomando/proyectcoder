*PORMPT UTILIZADO*

"Un joven enamorado de cabellos oscuros y vestimenta renacentista se encuentra en un jardín iluminado por la luz de la luna. Sus ojos reflejan devoción mientras observa a su amada en un balcón alto, con un vestido claro que ondea con la brisa nocturna. La escena es romántica y etérea, con un cielo estrellado y pétalos de rosa flotando en el aire, transmitiendo la intensidad de un amor prohibido."

```python
prompt = "A young man in Renaissance attire with dark hair stands in a moonlit garden, his eyes reflecting devotion as he gazes up at his beloved on a high balcony. She wears a light dress that flows gently in the night breeze. The scene is romantic and ethereal, with a starry sky and rose petals floating in the air, conveying the intensity of forbidden love."
```

```python
url = "https://api.stability.ai/v2beta/stable-image/generate/ultra"
headers = {
    "authorization": f"Bearer {API_KEY}",
    "accept": "image/*"
}
params = {
    "prompt": prompt,
    "output_format": "png"
}
```

```python
response = requests.post(
    url,
    headers=headers,
    files={"none": ''},
    data=params
)
```

***3. Aplicación y Ensayo***
- El actor usa los prompts para estudiar y ensayar.
- Usa imágenes generadas para reforzar la memoria visual

```python

print(response.status_code)
if response.status_code == 200:

    with open("imagen.png", 'wb') as file:
        file.write(response.content)
    print("✅ Imagen guardada como 'imagen.png'")


    from IPython.display import Image
    display(Image(response.content))
else:
    error_msg = response.json().get("message", "Error desconocido")
    raise Exception(f"❌ Error {response.status_code}: {error_msg}")
```
