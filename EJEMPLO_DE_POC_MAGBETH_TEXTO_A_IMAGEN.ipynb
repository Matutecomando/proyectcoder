```python
prompt = "A dark and somber scene with Macbeth staring at a floating dagger, his face filled with doubt and inner conflict. The atmosphere is tense, with deep shadows and dim lighting highlighting the anguish in his eyes. The background is blurred, evoking a sense of despair and doom."
```

```python
url = "https://api.stability.ai/v2beta/stable-image/generate/ultra"
headers = {
    "authorization": f"Bearer {API_KEY}",
    "accept": "image/*"
}
params = {
    "prompt": prompt,
    "output_format": "png"
}
```

```python
response = requests.post(
    url,
    headers=headers,
    files={"none": ''},
    data=params
)
```

```python

print(response.status_code)
if response.status_code == 200:

    with open("imagen.png", 'wb') as file:
        file.write(response.content)
    print("✅ Imagen guardada como 'imagen.png'")


    from IPython.display import Image
    display(Image(response.content))
else:
    error_msg = response.json().get("message", "Error desconocido")
    raise Exception(f"❌ Error {response.status_code}: {error_msg}")
```
